```python
from fduai.autograd import DataNode, nn, compile_nn, linear, compile_backward
from fduai.compiler import Compiler, Instruction, Variable
from fduai.compiler import generate_mlir
from fduai.common import Operator
import numpy as np

np.random.seed(42)
```

```python
with Compiler() as c:
    x = DataNode.from_list([2, 2], [[1, 2], [3, 4]])
    # add return statement manually so that this code can be compiled
    c.add_ret_stmt(None)

print(c.instructions)
```

```python
ir = generate_mlir(compiler=c)

with open('init.mlir', 'w') as fobj:
    fobj.write('/// example of memref initialization\n')
    fobj.write('/// Generated by initialization.ipynb\n')
    fobj.write(ir)
```

```python
with Compiler() as c:
    x = DataNode.uniform([2,3], -1.0, 1.0)
    c.add_ret_stmt(None)

for insn in c.instructions:
    print(insn)
```

```python
ir = generate_mlir(compiler=c)

with open('uniform.mlir', 'w') as fobj:
    fobj.write('/// example of memref initialization via uniform()\n')
    fobj.write('/// Generated by initialization.ipynb\n')
    fobj.write(ir)
```
